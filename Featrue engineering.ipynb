```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df.drop(['Row ID','Order ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Order Day'] = df['Order Date'].str.split('/').str[0]
df['Order Month'] = df['Order Date'].str.split('/').str[1]
df['Order Year'] = df['Order Date'].str.split('/').str[2]
```

```python
df['Order Day'] = df['Order Day'].astype(int)
df['Order Month'] = df['Order Month'].astype(int)
df['Order Year'] = df['Order Year'].astype(int)
```

```python
df.head()
```

```python
df['Ship Day'] = df['Ship Date'].str.split('/').str[0]
df['Ship Month'] = df['Ship Date'].str.split('/').str[1]
df['Ship Year'] = df['Ship Date'].str.split('/').str[2]
```

```python
df['Ship Day'] = df['Ship Day'].astype(int)
df['Ship Month'] = df['Ship Month'].astype(int)
df['Ship Year'] = df['Ship Year'].astype(int)
```

```python
df.head()
```

```python
df.drop(['Order Date','Ship Date'],axis=1 ,inplace=True)
```

```python
df.head()
```

```python
df['Ship Mode'].unique()
```

```python
df['Ship Mode'] = df['Ship Mode'].map({'Second Class':3, 'Standard Class':2, 'First Class':1, 'Same Day':0})
```

```python
df['Ship Mode'] = df['Ship Mode'].astype(int)
```

```python
df['Segment'].unique()
```

```python
df['Segment'] = df['Segment'].map({'Consumer':0, 'Corporate':1, 'Home Office':2})
```

```python
df['Country'].unique()
```
