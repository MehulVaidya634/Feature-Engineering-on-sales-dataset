# Import libraries and the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('train.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['Row ID','Order ID'],axis=1,inplace=True)
```

# Creating day month and year columns.

```python
df['Order Day'] = df['Order Date'].str.split('/').str[0]
df['Order Month'] = df['Order Date'].str.split('/').str[1]
df['Order Year'] = df['Order Date'].str.split('/').str[2]
```

```python
df['Order Day'] = df['Order Day'].astype(int)
df['Order Month'] = df['Order Month'].astype(int)
df['Order Year'] = df['Order Year'].astype(int)
```

```python
df.head()
```

# Creating shipping date, month and year.

```python
df['Ship Day'] = df['Ship Date'].str.split('/').str[0]
df['Ship Month'] = df['Ship Date'].str.split('/').str[1]
df['Ship Year'] = df['Ship Date'].str.split('/').str[2]
```

```python
df['Ship Day'] = df['Ship Day'].astype(int)
df['Ship Month'] = df['Ship Month'].astype(int)
df['Ship Year'] = df['Ship Year'].astype(int)
```

```python
df.head()
```

```python
df.drop(['Order Date','Ship Date'],axis=1 ,inplace=True)
```

```python
df.head()
```

# What was the best month for sales in 2018? How much was earned that month?

```python
new_df=df[df['Order Year']==2018]
```

```python
new_df.groupby('Order Month').sum().sort_values('Sales',ascending=False)
```

Answer : November month had the best sales in year 2018.
         the earned amount was  $117,938.1550 in November

# What State had the highest number of sales?

```python
sales_df = new_df.groupby('State').sum()
sales_df
```

```python
sales_df = sales_df.sort_values('Sales', ascending = False)
sales_df.head(5)
```

Answer : California has the most sales of amount $144666.7685

# What product was sold the most?

```python
product_df = new_df.groupby('Product Name').sum()
```

```python
product_df.sort_values('Sales',ascending=False).head()
```

Answer : Canon imageCLASS 2200 Advanced Copier was sold the most in 2018.
